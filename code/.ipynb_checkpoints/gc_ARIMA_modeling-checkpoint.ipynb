```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller

import warnings # necessary b/c pandas & statsmodels datetime issue
warnings.simplefilter(action="ignore")
```

```python
df = pd.read_csv('../data/gc_data/2019_04_09_22_21_05.csv')
```

```python
df.head(3)
```

## Data Cleaning

```python
# Convert index to datetime format
df.set_index(pd.to_datetime(df['secs']), inplace= True)
```

```python
# Drop the secs, km, and alt columns
df.drop(labels= ['secs', 'km', 'alt'], axis= 1, inplace= True)
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

## Smoothing

```python
df = df.rolling(20).mean();    # 20 sec rolling average
```

## Identify and correct trend

```python
df['hr'].plot(); # There is definitely a trend that needs to be handled
```

```python
pct_df = df.pct_change()
```

```python
pct_df.isnull().sum()
```

```python
# Replace NaN and inf with 0
pct_df = pct_df.fillna(0)
pct_df = pct_df.replace(np.inf, 0)
```

```python
pct_df.dtypes
```

```python
pct_df.isnull().sum()
```

```python
# Now we can see that the trend has been removed
pct_df['hr'].plot();
```

### PACF, ACF

```python
plot_acf(df['hr'], lags= 120);
```

```python
plot_pacf(df['hr'], lags= 10);
```

```python
plot_acf(pct_df['hr'], lags= 40);
```

```python
plot_pacf(pct_df['hr'], lags= 40);
```

## Train, Test, Split

```python
#y_train, y_test = train_test_split(pct_df['hr'], test_size=.2, shuffle=False)
y_train, y_test = train_test_split(pct_df['hr'], test_size=.2, shuffle=False)
```

#### Evaluating Stationarity

```python
def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput

interpret_dftest(adfuller(pct_df['hr']))
```

## Modeling

```python
# Instantiate
ar3 = ARIMA(endog= y_train, order= (2, 1, 2))

# Fit
ar3_model = ar3.fit()
```

```python
# Train data predictions
ar3_train_preds = ar3_model.predict(start= y_train.index[0], end= y_train.index[-1])

# Test data predictions
ar3_test_preds = ar3_model.predict(start= y_test.index[0], end= y_test.index[-1])
```

```python
mean_squared_error(y_train, ar3_train_preds), mean_squared_error(y_test, ar3_test_preds)
```

```python
# train, test r2 scores
r2_score(y_train, ar3_train_preds), r2_score(y_test, ar3_test_preds)
```

```python
def plot_results(train, test, preds):
    plt.figure(figsize=(14,6))

    # y_train
    plt.plot(train.index, pd.DataFrame(train), color = 'tab:blue', label='y_train')

    # y_test
    plt.plot(test.index, pd.DataFrame(test), color = 'tab:orange', label='y_test')

    # y_test predictions
    plt.plot(test.index, preds, color = 'teal', label='y_test predictions')

    plt.legend()
    plt.show();
    return None 
```

```python
plot_results(y_train, y_test, ar3_test_preds)
```

```python
# Instantiate
ar4 = ARIMA(endog= y_train, order= (52, 0, 1))

# Fit
ar4_model = ar4.fit()
```

```python
# Train data predictions
ar4_train_preds = ar4_model.predict(start= y_train.index[0], end= y_train.index[-1])

# Test data predictions
ar4_test_preds = ar4_model.predict(start= y_test.index[0], end= y_test.index[-1])
```

```python
mean_squared_error(y_train, ar4_train_preds), mean_squared_error(y_test, ar4_test_preds)
```

```python
# train, test r2 scores
r2_score(y_train, ar4_train_preds), r2_score(y_test, ar4_test_preds)
```

```python
plot_results(y_train, y_test, ar4_test_preds)
```
