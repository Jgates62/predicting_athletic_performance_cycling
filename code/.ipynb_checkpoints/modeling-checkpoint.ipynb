```python
# Add a negligible amount to columns that contain 0 values in order to prevent NaNs when using pct_change
df1_inc['Power'] = df1_inc['Power'] + 0.0001
df1_inc['Cadence'] = df1_inc['Cadence'] + 0.0001

# Set 'time' as the index
df1_inc.set_index('time', inplace= True)

# This should get rid of trend
hr_changes = df1_inc.pct_change()
```

### Train, Test, Split, and Modeling

```python
df1_inc
```

```python
hr_changes
```

```python
hr_changes.dtypes
```

```python
hr_changes.isnull().sum()
```

```python
hr_changes.dropna(inplace= True)
```

```python
# Create X and y
X = hr_changes.drop(columns= 'HR')
y = hr_changes['HR'].values
```

```python
X
```

```python
# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle= False)
```

```python
X_train
```

```python
# Scale
ss = StandardScaler()
Z_train = ss.fit_transform(X_train)
Z_test = ss.fit_transform(X_test)
```

```python
# Create training sequences
train_sequences = TimeseriesGenerator(Z_train, y_train, length=3, batch_size=64)

# Create test sequences
test_sequences = TimeseriesGenerator(Z_test, y_test, length=3, batch_size=64)
```

```python
Z_train[0].shape
```

```python
# RNN Design ** Need to know what the input shape should actually be, and why - probably review the RNN vido lecture

model = Sequential()

model.add(GRU(4, input_shape=(3, 4), return_sequences=True)) # True if the next layer is RNN
model.add(GRU(4, return_sequences=False)) # False if the next layer is Dense
model.add(Dense(4, activation='relu'))
# output layer for binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse', optimizer=Adam(lr=.0005), metrics= 'mse')
```

```python
res = model.fit(train_sequences, epochs=100, validation_data=test_sequences)
```

```python
plt.plot(res.history['loss'], label= 'Train')
plt.plot(res.history['val_loss'], label= 'Test');
```

```python
# Train
r2_score(y_train, model.predict(Z_train))
```

```python
model.predict(Z_train)
```

```python
# Test
r2_score(y_test, model.predict(Z_test))
```
