```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/gc_data/2019_04_09_22_21_05.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.rolling(5).mean();    # 5 sec rolling average
```

```python
# Plot the data
sns.set_style('darkgrid')

plt.figure(figsize = (16, 6))
plt.title('Heart Rate, Power * 0.5, Cadence', fontsize= 14)


x = df['secs']

ya = df['hr']
yb = df['power'] * 0.5
yc = df['cad']

plt.plot(x, ya, c = 'tab:red', label= 'Heart Rate (bpm)')
plt.plot(x, yb, c = 'tab:blue', label= 'Power (watts * 0.5)')
plt.plot(x, yc, c = 'tab:green', label= 'Pedaling Cadence (cycles/minute)')

plt.ylim(75, 180)

plt.legend();
```
