```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_info = pd.read_csv('../data/data_int/test_subj_info_kag_cycling.csv')
df_info
```

```python
# PPO : Peak Power Output
```

```python
df1_inc = pd.read_csv('../data/data_int/sbj_1_incremental.csv')
df1_inc
```

```python
df_vis = df1_inc['Oxygen'] * 0.1
df_vis
```

```python
# Plot the data
sns.set_style('darkgrid')

plt.figure(figsize = (16, 6))
plt.title('Athlete 1 Incremental Test: Heart Rate, Power * 0.5, Cadence', fontsize= 14)


x = df1_inc['time']

ya = df1_inc['HR']
yb = (df1_inc['Power'] * 0.5)
yc = df1_inc['Cadence']

plt.plot(x, ya, c = 'tab:red', label= 'Heart Rate (bpm)')
plt.plot(x, yb, c = 'tab:blue', label= 'Power (watts * 0.5)')
plt.plot(x, yc, c = 'tab:green', label= 'Pedaling Cadence (cycles/minute)')

plt.legend();
```
