```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_info = pd.read_csv('../data/data_int/test_subj_info_kag_cycling.csv')
df_info
```

```python
# PPO : Peak Power Output
```

```python
df1_inc = pd.read_csv('../data/data_int/sbj_1_incremental.csv')
df1_inc
```

```python
df_vis = df1_inc['Oxygen'] * 0.1
df_vis
```

```python
# Plot the data
sns.set_style('darkgrid')

plt.figure(figsize = (20, 8))

ax1 = plt.subplot(2, 1, 1, title= 'HR and RF', ylabel= 'HR and RF')
ax2 = plt.subplot(2, 1, 2, title= 'Power and Oxygen Consumption')

x = df1_inc['time']

y1 = df1_inc['HR']
y2 = df1_inc['RF']

ya = df1_inc['Power']
yb = (df1_inc['Oxygen'] * 0.1)
yc = df1_inc['Cadence']


ax1.plot(x, y1, c = 'tab:red', label= 'Heart Rate')
ax1.plot(x, y2, c = 'tab:blue', label= 'Respiratory Frequency')

ax2.plot(x, ya, c = 'tab:orange', label= 'Power in Watts')
ax2.plot(x, yb, c = 'teal', label= 'Oxygen Consumption * 0.1')
ax2.plot(x, yc, c = 'tab:brown', label= 'Cadence')

ax1.legend();
ax2.legend();
```
