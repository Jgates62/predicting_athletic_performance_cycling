```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1_inc = pd.read_csv('../data/data_int/sbj_1_incremental.csv')
df1_win = pd.read_csv('../data/data_int/sbj_1_Wingate.csv')
df1_t1 = pd.read_csv('../data/data_int/sbj_1_I.csv')
df1_t2 = pd.read_csv('../data/data_int/sbj_1_II.csv')
```

```python
# Create a feature called fatigue that is the difference between power and HR
```

```python
df1_t1['fatigue'] = df1_t1['HR'] - (df1_t1['Power'] * 0.25)
#df1_t1['fatigue'] = df1_t1['Power'] - df1_t1['HR']
```

```python
plt.figure(figsize= (14, 5))
plt.plot(df1_t1['HR'], color = 'tab:red')
plt.plot(df1_t1['Power'] * 0.5, color= 'tab:blue')
plt.plot(df1_t1['fatigue'], color= 'green');
```
