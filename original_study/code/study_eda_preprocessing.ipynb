## Data Visualization

```python
def plot_study_workout(cur_df, cur_title): # Takes the current dataframe, title as input and visualizes the data
    
    sns.set_style('darkgrid')
    plt.figure(figsize= (16, 5))
    plt.title(cur_title, fontsize= 15)
    plt.plot(cur_df['HR'], color = 'tab:red', label= 'Heart Rate (BPM)')
    plt.plot(cur_df['Power'] * 0.5, color= 'tab:blue', label= 'Power (watts) * 0.5')
    plt.plot(cur_df['Cadence'], color= 'tab:green', label= 'Pedaling Cadence')
    plt.plot(cur_df['Oxygen'] * 0.1, color= 'tab:orange', label= 'Oxygen Consumption/10')
    plt.xlabel('Time in Seconds', fontsize= 12)
    plt.legend()
    plt.savefig(f'../visualizations/{cur_title}.jpg', bbox_inches='tight');
    
    return None
```

```python
plot_study_workout(df1_inc, 'Original Study Incremental Test')
```

```python
plot_study_workout(df1_win, 'Original Study Wingate Test')
```

```python
plot_study_workout(df1_t1, 'Original Study Test 1')
```

```python
plot_study_workout(df1_t2, 'Original Study Test 2')
```

## Split workouts up between train and test data

```python
# The Incremental, Wingate, and Test 1 dataframes will be used for training the model
train = pd.concat([df1_inc, df1_win, df1_t1], ignore_index= True)
```

```python
# The Test 2 dataframe will be used to test the model
test = df1_t2
```

```python
plot_study_workout(train, 'Original Study Training Data')
```

```python
plot_study_workout(test, 'Original Study Testing Data')
```
